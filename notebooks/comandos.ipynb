## Comandos importantes (pandas)

### 1. Limpeza de Dados (Data Cleaning) 

- Verificar dados ausentes (NaN):

```python
df.isnull().sum()
```

- Remover linhas com qualquer valor nulo:

```python
df.dropna()
```

- Remover colunas com valores nulos: 

```python
df.dropna(axis=1)
```

- Preencher valores nulos: 

```python
df.fillna(valor) ou df.fillna(df.mean())
```

- Remover linhas duplicadas:

```python
df.drop_duplicates()
```

- Renomear colunas:

```python
df.rename(columns={'velho': 'novo'})
```

- Alterar o tipo de dado: 

```python
df['coluna'].astype(tipo) (ex: int, float, str)
```

- Remover espaços em branco (strings):

```python
df['coluna'].str.strip()
```

- Converter para minúsculas: 

```python
df['coluna'].str.lower()
```

- Substituir valores específicos:

```python
df.replace('valor_antigo', 'valor_novo')
```

### 2. Seleção e Filtragem de Dados (Data Selection)

- Selecionar uma coluna:

```python
df['nome_coluna']
```

- Selecionar múltiplas colunas:

```python
df[['coluna1', 'coluna2']]
```

- Selecionar linhas por posição (índice):

```python
df.iloc[0:5] #(primeiras 5 linhas)
```

- Selecionar linhas e colunas por posição: 

```python
df.iloc[0:5, 0:3]
```

- Selecionar por rótulo (Label):

```python
df.loc[linhas, colunas]
```

- Filtrar com condições (Boolean Indexing): 

```python
df[df['coluna'] > 10]
```

- Filtrar com múltiplas condições:

```python
df[(df['coluna1'] > 10) & (df['coluna2'] == 'A')]
```

- Filtrar usando .query(): 

```python
df.query('coluna > 10')
```

- Filtrar strings contendo um valor: 

```python
df[df['coluna'].str.contains('texto')]
```

- Selecionar valores únicos:

```python
df['coluna'].unique()
```
