```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
url = 'https://tinyurl.com/spothits'
df = pd.read_csv(url)
df.head()
```

```python
df.isnull().sum()
```

```python
#Apagar dados faltantes
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
# Retornando as músicas mais populares
df.nlargest(10, ["track_popularity"]) #repete a mesma música mas com o id diferente

# os dados estão duplicados pq pode ter musicas igauis em albuns diferentes
```

```python
# Removendo dados duplicados
df_limpo = df.drop_duplicates(["track_name"])
```

```python
df.size
```

```python
display(df_limpo.size)
```

```python
df_limpo.track_popularity.hist()
```

```python
sns.histplot(df['track_popularity'], kde=True)
```

```python
df_limpo.playlist_genre.value_counts()
```

```python
#plotando um gráfico da distribuição dos gêneros musicais
contagem = df_limpo['playlist_genre'].value_counts()

plt.figure(figsize=(12,6))
sns.barplot(x=contagem.index ,y=contagem.values, palette="spring")
plt.xlabel("Gênero da playlist")
plt.ylabel("Contagem")
plt.title("Distribuição dos Gêneros de Playlist")
plt.show()
```

```python
#Obtendo o BOX PLOT de Popularidade
plt.figure(figsize=(10,2))
sns.boxplot(x=df_limpo['track_popularity'])
```

```python
plt.figure(figsize=(10,5))
sns.histplot(df_limpo["danceability"], kde=True)
plt.title("Distribuição de Danceability")
plt.show()
```

### Filtrando o dataframe

```python
df_hit=df_limpo[df_limpo.track_popularity <= 58 ]
df_naohit = df_limpo[ df_limpo.track_popularity <= 23 ]
```

```python
print(df_hit.shape)
print(df_naohit.shape)
```

```python
plt.figure(figsize=(10,5))
sns.hisplot(df_hit["dance"])
```

```python
# Fluxo completo para mais uma variável
variaveis_alvo = ['track_id', 'track_name', 'track_artist', 'track_popularity',
       'track_album_id', 'track_album_name', 'track_album_release_date',
       'playlist_name', 'playlist_id', 'playlist_genre', 'playlist_subgenre',
       'danceability', 'energy', 'key', 'loudness', 'mode', 'speechiness',
       'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo',
       'duration_ms']

# para cada variavel, plotar os gráficos para HIT e NÃO HIT
for variavel in variaveis_alvo:
    plt.figure(figsize=(10,5))
    sns.histplot(df_hit[variavel], kde=True, color="skyblue")
    plt.title(f"distribuição de {variavel} para HITS")
    plt.show()

    plt.figure(figsize=(10,5))
    sns.histplot(df_naohit[variavel], kde=True, color="skyble")
    plt.title(f"distribuição de {variavel} para NÃO HITS")
    plt.show()
```
